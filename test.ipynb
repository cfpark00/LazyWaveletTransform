```python
import LWT
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
wavelets=LWT.make_wavelets(N=256,NR=6,NT=8)
wavelets.shape
```

```python
im=torch.randn(256,256,dtype=torch.float32)
```

```python
res=LWT.LWT_R(im,wavelets,rsnl="logabsp1",m=2)
res.shape
```

```python
plt.figure(figsize=(20,6))
plt.plot(res)
plt.yscale("log")
```
